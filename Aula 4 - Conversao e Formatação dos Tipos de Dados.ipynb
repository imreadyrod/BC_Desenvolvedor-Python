Conversão dos tipos de Dados

```python
nascimento = 1986
ano_atual = 2022
idade = ano_atual - nascimento
print('Eu tenho', idade, 'anos')
```

```python
saida = 'Eu tenho ' + str(idade) + ' anos'
print(saida)
```

```python
saida2 = str(idade)
print('saida2' -3 )
```

```python
#conversão do tipo int
print(float(10))
print(bool(-1))
print(bool(0))
print(str(-999))
```

```python
#conversão do tipo str
print(int('-99'))
print(float('0.01'))
print(bool('palavra'))
```
