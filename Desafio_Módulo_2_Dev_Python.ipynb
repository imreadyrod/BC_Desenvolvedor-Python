```python
#QH - Qual estação do ano contém a maior média de locações de bicicletas? Estação 3 = verão
#QI - Qual estação do ano contém a menor média de locações de bicicletas? Estação 1 = inverno
colunasH = ['season','total_count']
questaoH = df.filter(items=colunasH) 
```

```python
questaoH.head()
```

```python
questaoH.groupby('season').count()
```

```python
#Qj - Qual horário do dia contém a maior média de locações de bicicletas?  17 horas
#QK - Qual horário do dia contém a menor média de locações de bicicletas?  4 horas
colunasJK = ['hour','total_count']
questaoJK = df.filter(items=colunasJK) 
```

```python
questaoJK.head()
```

```python
questaoJK.groupby('hour').mean()
```

```python
questaoJK.groupby('hour').max()
```

```python
# l. Que dia da semana contém a maior média de locações de bicicletas? dia 4 = quinta feira 
# m. Que dia da semana contém a menor média de locações de bicicletas? dia 0 = domingo
colunasLM = ['weekday','total_count']
questaoLM = df.filter(items=colunasLM) 
```

```python
questaoLM.head()
```

```python
questaoLM.groupby('weekday').mean()
```
